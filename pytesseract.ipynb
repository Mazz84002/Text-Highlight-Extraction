## OpenCV

```python
import cv2
```

```python
img_cv = cv2.imread('/Users/mazz/Documents/Programming/Python/Spacy/data/text for pytesseract.png')
```

```python
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
```

```python
print(pytesseract.image_to_string(img_rgb))
```

```python
print(pytesseract.image_to_data(img_rgb))
```

```python
cv2.imshow('MY_TEXT', img_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# The --oem argument, or OCR Engine Mode, controls the type of algorithm used by Tesseract.
# The --psm controls the automatic Page Segmentation Mode used by Tesseract.


# Example of adding any additional options
custom_oem_psm_config = r'--oem 3 --psm 6'
pytesseract.image_to_string(img_rgb, config=custom_oem_psm_config)
```
